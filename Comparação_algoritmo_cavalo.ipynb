**BACKTRACKING**

```python
import numpy as np
import time

def movimento_valido(x, y, tabuleiro, tamanho):
    return 0 <= x < tamanho and 0 <= y < tamanho and tabuleiro[x][y] == -1

def resolver_passeio(tabuleiro, x, y, contador_movimentos, movimentos_x, movimentos_y, tamanho, passos, ordem_passos):
    if contador_movimentos == tamanho * tamanho:
        return True, passos + 1

    for i in range(8):
        proximo_x, proximo_y = x + movimentos_x[i], y + movimentos_y[i]
        if movimento_valido(proximo_x, proximo_y, tabuleiro, tamanho):
            tabuleiro[proximo_x][proximo_y] = contador_movimentos
            ordem_passos.append((proximo_x, proximo_y))
            sucesso, passos = resolver_passeio(tabuleiro, proximo_x, proximo_y, contador_movimentos + 1, movimentos_x, movimentos_y, tamanho, passos + 1, ordem_passos)
            if sucesso:
                return True, passos
            tabuleiro[proximo_x][proximo_y] = -1  # Backtracking
            ordem_passos.pop()

    return False, passos

def exibir_tabuleiro(tabuleiro):
    print("Tabuleiro final:")
    for linha in tabuleiro:
        print(" ".join(f"{num:2}" for num in linha))

def passeio_do_cavalo(tamanho, inicio_x, inicio_y):
    tabuleiro = np.full((tamanho, tamanho), -1)

    # Movimentos possíveis do cavalo
    movimentos_x = [2, 1, -1, -2, -2, -1, 1, 2]
    movimentos_y = [1, 2, 2, 1, -1, -2, -2, -1]

    tabuleiro[inicio_x][inicio_y] = 0  # Define a posição inicial
    ordem_passos = [(inicio_x, inicio_y)]

    tempo_inicio = time.time()
    sucesso, passos = resolver_passeio(tabuleiro, inicio_x, inicio_y, 1, movimentos_x, movimentos_y, tamanho, 0, ordem_passos)
    tempo_fim = time.time()

    if not sucesso:
        print("Não foi possível encontrar uma solução.")
    else:
        exibir_tabuleiro(tabuleiro)
        print(f"Tempo de execução: {tempo_fim - tempo_inicio:.6f} segundos")
        print(f"Quantidade de passos: {passos}")
        print("Ordem dos movimentos:")
        for i, (linha, coluna) in enumerate(ordem_passos):
            print(f"Passo {i}: ({linha}, {coluna})")

if __name__ == "__main__":
    tamanho = 8  # Tamanho do tabuleiro
    entrada = input("Digite a posição inicial do cavalo no formato 'linha coluna' (exemplo: 0 0): ")
    inicio_x, inicio_y = map(int, entrada.split())
    passeio_do_cavalo(tamanho, inicio_x, inicio_y)
```

**TENTATIVA E ERRO PURO**

```python
import numpy as np
import random
import time

def movimento_valido(x, y, tabuleiro, tamanho):
    return 0 <= x < tamanho and 0 <= y < tamanho and tabuleiro[x][y] == -1

def passeio_do_cavalo_tentativa_erro(tamanho, inicio_x, inicio_y):
    while True:  # Continua tentando até encontrar uma solução completa
        tabuleiro = np.full((tamanho, tamanho), -1)
        tabuleiro[inicio_x][inicio_y] = 0  # Define a posição inicial

        movimentos_x = [2, 1, -1, -2, -2, -1, 1, 2]
        movimentos_y = [1, 2, 2, 1, -1, -2, -2, -1]

        x, y = inicio_x, inicio_y
        ordem_passos = [(x, y)]
        movimentos_realizados = 1

        tempo_inicio = time.time()

        while movimentos_realizados < tamanho * tamanho:
            movimentos_possiveis = []
            for i in range(8):
                proximo_x = x + movimentos_x[i]
                proximo_y = y + movimentos_y[i]
                if movimento_valido(proximo_x, proximo_y, tabuleiro, tamanho):
                    movimentos_possiveis.append((proximo_x, proximo_y))

            if not movimentos_possiveis:
                # Se ficar preso, reinicia o tabuleiro do zero
                break

            # Escolhe um movimento aleatório
            x, y = random.choice(movimentos_possiveis)
            tabuleiro[x][y] = movimentos_realizados
            ordem_passos.append((x, y))
            movimentos_realizados += 1

        tempo_fim = time.time()

        if movimentos_realizados == tamanho * tamanho:
            # Solução encontrada
            print("Solução encontrada!")
            for linha in tabuleiro:
                print(" ".join(f"{num:2}" for num in linha))
            print(f"Tempo de execução: {tempo_fim - tempo_inicio:.6f} segundos")
            print(f"Quantidade de passos: {movimentos_realizados}")
            print("Ordem dos movimentos:")
            for i, (linha, coluna) in enumerate(ordem_passos):
                print(f"Passo {i}: ({linha}, {coluna})")
            break  # Sai do loop infinito ao encontrar uma solução

if __name__ == "__main__":
    tamanho = 8  # Tamanho do tabuleiro
    entrada = input("Digite a posição inicial do cavalo no formato 'linha coluna' (exemplo: 0 0): ")
    inicio_x, inicio_y = map(int, entrada.split())
    passeio_do_cavalo_tentativa_erro(tamanho, inicio_x, inicio_y)
```

**PRIORIZANDO CANTOS MAIS ABERTOS**

```python
import numpy as np
import time

def movimento_valido(x, y, tabuleiro, tamanho):
    return 0 <= x < tamanho and 0 <= y < tamanho and tabuleiro[x][y] == -1

def distancia_para_canto(x, y, tabuleiro, tamanho):
    # Distância normal até os cantos do tabuleiro
    distancia_canto = min(x, y, tamanho - 1 - x, tamanho - 1 - y)

    # Conta quantas casas ao redor já foram visitadas
    casas_visitadas = 0
    movimentos_x = [2, 1, -1, -2, -2, -1, 1, 2]
    movimentos_y = [1, 2, 2, 1, -1, -2, -2, -1]

    for i in range(8):
        proximo_x, proximo_y = x + movimentos_x[i], y + movimentos_y[i]
        if 0 <= proximo_x < tamanho and 0 <= proximo_y < tamanho:
            if tabuleiro[proximo_x][proximo_y] != -1:
                casas_visitadas += 1

    # Penaliza a distância se estiver cercado por muitas casas visitadas
    return distancia_canto + casas_visitadas

def resolver_passeio(tabuleiro, x, y, contador_movimentos, movimentos_x, movimentos_y, tamanho, passos, ordem_passos):
    if contador_movimentos == tamanho * tamanho:
        return True, passos + 1

    # Movimentos ordenados pela proximidade de um canto + consideração de casas visitadas
    movimentos_possiveis = []
    for i in range(8):
        proximo_x, proximo_y = x + movimentos_x[i], y + movimentos_y[i]
        if movimento_valido(proximo_x, proximo_y, tabuleiro, tamanho):
            distancia = distancia_para_canto(proximo_x, proximo_y, tabuleiro, tamanho)
            movimentos_possiveis.append((distancia, proximo_x, proximo_y))

    # Ordena para priorizar movimentos que nos levam a espaços mais "seguros"
    movimentos_possiveis.sort()

    for _, proximo_x, proximo_y in movimentos_possiveis:
        tabuleiro[proximo_x][proximo_y] = contador_movimentos
        ordem_passos.append((proximo_x, proximo_y))
        sucesso, passos = resolver_passeio(tabuleiro, proximo_x, proximo_y, contador_movimentos + 1, movimentos_x, movimentos_y, tamanho, passos + 1, ordem_passos)
        if sucesso:
            return True, passos
        tabuleiro[proximo_x][proximo_y] = -1  # Backtracking
        ordem_passos.pop()

    return False, passos

def exibir_tabuleiro(tabuleiro):
    print("Tabuleiro final:")
    for linha in tabuleiro:
        print(" ".join(f"{num:2}" for num in linha))

def passeio_do_cavalo(tamanho, inicio_x, inicio_y):
    tabuleiro = np.full((tamanho, tamanho), -1)

    # Movimentos possíveis do cavalo
    movimentos_x = [2, 1, -1, -2, -2, -1, 1, 2]
    movimentos_y = [1, 2, 2, 1, -1, -2, -2, -1]

    tabuleiro[inicio_x][inicio_y] = 0  # Define a posição inicial
    ordem_passos = [(inicio_x, inicio_y)]

    tempo_inicio = time.time()
    sucesso, passos = resolver_passeio(tabuleiro, inicio_x, inicio_y, 1, movimentos_x, movimentos_y, tamanho, 0, ordem_passos)
    tempo_fim = time.time()

    if not sucesso:
        print("Não foi possível encontrar uma solução.")
    else:
        exibir_tabuleiro(tabuleiro)
        print(f"Tempo de execução: {tempo_fim - tempo_inicio:.6f} segundos")
        print(f"Quantidade de passos: {passos}")
        print("Ordem dos movimentos:")
        for i, (linha, coluna) in enumerate(ordem_passos):
            print(f"Passo {i}: ({linha}, {coluna})")

if __name__ == "__main__":
    tamanho = 8  # Tamanho do tabuleiro
    entrada = input("Digite a posição inicial do cavalo no formato 'linha coluna' (exemplo: 0 0): ")
    inicio_x, inicio_y = map(int, entrada.split())
    passeio_do_cavalo(tamanho, inicio_x, inicio_y)
```

**(Desconsiderar) - warnsdorff para comparação**

```python
import numpy as np
import time

def movimento_valido(x, y, tabuleiro, tamanho):
    """Verifica se o movimento está dentro do tabuleiro e se a casa não foi visitada."""
    return 0 <= x < tamanho and 0 <= y < tamanho and tabuleiro[x][y] == -1

def movimentos_disponiveis(x, y, tabuleiro, tamanho):
    """Conta quantos movimentos válidos existem a partir da posição (x, y)."""
    movimentos_x = [2, 1, -1, -2, -2, -1, 1, 2]
    movimentos_y = [1, 2, 2, 1, -1, -2, -2, -1]
    contador = 0

    for i in range(8):
        proximo_x, proximo_y = x + movimentos_x[i], y + movimentos_y[i]
        if movimento_valido(proximo_x, proximo_y, tabuleiro, tamanho):
            contador += 1

    return contador  # Retorna o número de movimentos futuros disponíveis

def resolver_passeio(tabuleiro, x, y, contador_movimentos, movimentos_x, movimentos_y, tamanho, passos, ordem_passos):
    """Resolve o passeio do cavalo usando a heurística de Warnsdorff."""
    if contador_movimentos == tamanho * tamanho:
        return True, passos + 1

    # Lista de movimentos ordenados pelo critério de Warnsdorff (menos opções futuras primeiro)
    movimentos_possiveis = []
    for i in range(8):
        proximo_x, proximo_y = x + movimentos_x[i], y + movimentos_y[i]
        if movimento_valido(proximo_x, proximo_y, tabuleiro, tamanho):
            grau_liberdade = movimentos_disponiveis(proximo_x, proximo_y, tabuleiro, tamanho)
            movimentos_possiveis.append((grau_liberdade, proximo_x, proximo_y))

    # Ordena para priorizar as posições com MENOS opções futuras (Warnsdorff)
    movimentos_possiveis.sort()

    for _, proximo_x, proximo_y in movimentos_possiveis:
        tabuleiro[proximo_x][proximo_y] = contador_movimentos
        ordem_passos.append((proximo_x, proximo_y))
        sucesso, passos = resolver_passeio(tabuleiro, proximo_x, proximo_y, contador_movimentos + 1, movimentos_x, movimentos_y, tamanho, passos + 1, ordem_passos)
        if sucesso:
            return True, passos
        tabuleiro[proximo_x][proximo_y] = -1  # Backtracking
        ordem_passos.pop()

    return False, passos

def exibir_tabuleiro(tabuleiro):
    """Exibe o tabuleiro final após a solução."""
    print("Tabuleiro final:")
    for linha in tabuleiro:
        print(" ".join(f"{num:2}" for num in linha))

def passeio_do_cavalo(tamanho, inicio_x, inicio_y):
    """Inicia o passeio do cavalo e mede o tempo de execução."""
    tabuleiro = np.full((tamanho, tamanho), -1)

    movimentos_x = [2, 1, -1, -2, -2, -1, 1, 2]
    movimentos_y = [1, 2, 2, 1, -1, -2, -2, -1]

    tabuleiro[inicio_x][inicio_y] = 0
    ordem_passos = [(inicio_x, inicio_y)]

    tempo_inicio = time.time()
    sucesso, passos = resolver_passeio(tabuleiro, inicio_x, inicio_y, 1, movimentos_x, movimentos_y, tamanho, 0, ordem_passos)
    tempo_fim = time.time()

    if not sucesso:
        print("Não foi possível encontrar uma solução.")
    else:
        exibir_tabuleiro(tabuleiro)
        print(f"Tempo de execução: {tempo_fim - tempo_inicio:.6f} segundos")
        print(f"Quantidade de passos: {passos}")

if __name__ == "__main__":
    tamanho = 8
    entrada = input("Digite a posição inicial do cavalo no formato 'linha coluna' (exemplo: 0 0): ")
    inicio_x, inicio_y = map(int, entrada.split())
    passeio_do_cavalo(tamanho, inicio_x, inicio_y)
```
